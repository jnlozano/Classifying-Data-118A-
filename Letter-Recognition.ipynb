```python
##-- Letter Recognition Data --##
import warnings
warnings.filterwarnings("ignore")
import sklearn.model_selection as cv
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn import svm

# X: Copy labeling from original paper
X = [list(map(int, x.split(',')[1:])) for x in open('letter-recognition.data').read().splitlines()]
Y = [1 if x.split(',')[0] == 'O' else -1 for x in open('letter-recognition.data').read().splitlines()]

xTrain, xTest, yTrain, yTest = cv.train_test_split(X, Y, train_size = 4000/float(len(X)))
xTrain50, xTest50, yTrain50, yTest50 = cv.train_test_split(X, Y, train_size = 10000/float(len(X)))
xTrain80, xTest80, yTrain80, yTest80 = cv.train_test_split(X, Y, train_size = 16000/float(len(X)))

import import_ipynb
import Methods as m

m.KNN(xTrain, xTest, yTrain, yTest)
m.RandomForest(xTrain, xTest, yTrain, yTest)
m.SVMlin(xTrain, xTest, yTrain, yTest)
m.SVMrbf(xTrain, xTest, yTrain, yTest)
m.BoostedDecisionTree(xTrain, xTest, yTrain, yTest)
m.NeuralNets(xTrain, xTest, yTrain, yTest)
m.XGBoost(xTrain, xTest, yTrain, yTest)
```

```python
print("50/50 Data")
m.KNN(xTrain50, xTest50, yTrain50, yTest50)
m.RandomForest(xTrain50, xTest50, yTrain50, yTest50)
m.SVMlin(xTrain50, xTest50, yTrain50, yTest50)
m.SVMrbf(xTrain50, xTest50, yTrain50, yTest50)
m.BoostedDecisionTree(xTrain50, xTest50, yTrain50, yTest50)
m.NeuralNets(xTrain50, xTest50, yTrain50, yTest50)
m.XGBoost(xTrain50, xTest50, yTrain50, yTest50)
```

```python
print("80/20 Data")
m.KNN(xTrain80, xTest80, yTrain80, yTest80)
m.RandomForest(xTrain80, xTest80, yTrain80, yTest80)
m.SVMlin(xTrain80, xTest80, yTrain80, yTest80)
m.SVMrbf(xTrain80, xTest80, yTrain80, yTest80)
m.BoostedDecisionTree(xTrain80, xTest80, yTrain80, yTest80)
m.NeuralNets(xTrain80, xTest80, yTrain80, yTest80)
m.XGBoost(xTrain80, xTest80, yTrain80, yTest80)
```

```python
classifier = svm.SVC(kernel='linear')   
params = {"C": [0.1, 1, 2, 3, 4, 5, 6, 7]}
GSC = GridSearchCV(classifier, params, return_train_score=True, cv=5, n_jobs=4)
GSC.fit(xTrain,yTrain)
results = GSC.cv_results_

train_acc = np.matrix(results.get("mean_train_score")).T
m.draw_heatmap_linear(train_acc, 'train accuracy', [0.1, 1, 2, 3, 4, 5, 6, 7])
val_acc = np.matrix(results.get("mean_test_score")).T
m.draw_heatmap_linear(val_acc, 'val accuracy', [0.1, 1, 2, 3, 4, 5, 6, 7])


classifier = svm.SVC(kernel='rbf')
params = {"C": [0.1, 1, 2, 3, 4, 5, 6, 7], "gamma": [500000.0, 20000.0, 5000.0, 200, 50, 2.0, 0.5, 0.125]}
GSR = GridSearchCV(classifier, params, return_train_score=True, cv=5, n_jobs=4)
GSR.fit(xTrain,yTrain)
```
