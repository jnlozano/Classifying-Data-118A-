```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
import sklearn.model_selection as cv
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn import svm


data = pd.read_csv('crx.data', sep=",", header=None)
data.columns = ["A1","A2","A3","A4","A5","A6","A7","A8","A9","A10","A11","A12","A13","A14","A15","A16"]
data = data.replace(' ','', regex=True)
data_cont = data.iloc[:,[1,2,7,10,13,14]]
data_cat = data.iloc[:,[0,3,4,5,6,8,9,11,12,15]]
data_dummies = pd.get_dummies(data_cat)
data_cont = data_cont.replace('?', 0)

d_s=StandardScaler()
d_n=d_s.fit_transform(data_cont) 

X_and_Y = np.hstack([d_n, data_dummies.values])

np.random.shuffle(X_and_Y)
X = X_and_Y[:5000, 0:-1]
Y = X_and_Y[:5000, -1]

split20 = int(round(len(X)*0.2))
split50 = int(round(len(X)*0.5))
split80 = int(round(len(X)*0.8))

xTrain = X[:split20,:] # Get features from train + val set.
xTest  = X[split20:len(X),:] # Get features from test set.     
yTrain = Y[:split20] 
yTest  = Y[split20:len(X)] 

xTrain50 = X[:split50,:] 
xTest50  = X[split50:len(X),:]     
yTrain50 = Y[:split50] 
yTest50  = Y[split50:len(X)] 

xTrain80 = X[:split80,:] 
xTest80  = X[split80:len(X),:]     
yTrain80 = Y[:split80] 
yTest80  = Y[split80:len(X)] 

import import_ipynb
import Methods as m

m.KNN(xTrain, xTest, yTrain, yTest)
m.RandomForest(xTrain, xTest, yTrain, yTest)
m.SVMlin(xTrain, xTest, yTrain, yTest)
m.SVMrbf(xTrain, xTest, yTrain, yTest)
m.BoostedDecisionTree(xTrain, xTest, yTrain, yTest)
m.NeuralNets(xTrain, xTest, yTrain, yTest)
m.XGBoost(xTrain, xTest, yTrain, yTest)
```

```python
classifier = svm.SVC(kernel='linear')   
params = {"C": [10, 1, 0.1, 0.01, 0.001, 0.0001, .00001,.000001]}
GSC = GridSearchCV(classifier, params, return_train_score=True, cv=5, n_jobs=4)
GSC.fit(xTrain,yTrain)
GSC.cv_results_
test_acc = sum(GSC.best_estimator_.predict(xTest) == yTest) / len(xTest)
results = GSC.cv_results_

train_acc = np.matrix(results.get("mean_train_score")).T
m.draw_heatmap_linear(train_acc, 'train accuracy', [0.1, 1, 2, 3, 4, 5, 6, 7])
val_acc = np.matrix(results.get("mean_test_score")).T
m.draw_heatmap_linear(val_acc, 'val accuracy', [0.1, 1, 2, 3, 4, 5, 6, 7])
```
